A Jupyter notebook is composed of cells. There are two types of cell: Markdown and Code. I'm writing this in a markdown cell. These are used for pieces of prose, and can essentially display anything that a webpage can (text, figures, videos...). If you try double-clicking on this cell, you'll be able to see the "source code". When you're done, press shift-enter and that will convert it back into the "rendered" version that displays nicely.

The Code cell type is where you're going to type code. These are set up with "syntax highlighting". Basically, because it knows you're going to be typing python code, and it knows what python code should look like, it'll colour what you write in a helpful way. It's often useful in picking up various common, easy to make, mistakes. Again, when you've written code in a code cell, hit shift-enter. This will run your code.

You can always go back and edit the code in a cell and run it again.

You can add new cells at any point. Click on the cell above where you want the new cell, and click the + button in the toolbar. (Try this now.)
